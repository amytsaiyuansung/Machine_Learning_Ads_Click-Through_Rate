# Ordinal Encoding for Faster Computation

```python
# Split x and y
xtrain = train021.iloc[ : , 1: ]
ytrain= train021.iloc[ :, 0]

xval = val.iloc[:, 1: ]
yval = val.iloc[:, 0]
```

```python
# Create a full dataset to fit so that the model does not include unseen labels
xtrainval = pd.concat([xval,xtrain])
```

```python
# Recode value into ordinal values
from sklearn.preprocessing import LabelEncoder

xtrainval_label = xtrainval.values
xtrain_label = xtrain.values
xval_label = xval.values

for i in range(0,21):
    le = LabelEncoder()
    le.fit(xtrainval.iloc[:, i])
    xtrain_label[:, i] = le.transform(xtrain_label[:, i])
    xval_label[:, i] = le.transform(xval_label[:, i])
```

```python
xtrain_label.shape
```

```python
# Change from numpy array to dataframe
xtrain_df = pd.DataFrame(xtrain_label)
xval_df = pd.DataFrame(xval_label)
```

## Combine Category

```python
# Combine levels for selected variables - Train
for i in np.r_[3,4,6,7,9,12,15,18]:
    count =pd.DataFrame(xtrain_df.iloc[:, i].value_counts())
    count['freq'] = count[count.columns[0]].cumsum()
    category = count.index.values[:(count[count.freq <= 800000].shape[0]+1)]
    allcategory =xtrain_df.iloc[:,i].unique()
    diff = np.setdiff1d(allcategory, category)
    xtrain_df.iloc[: ,i] = xtrain_df.iloc[:,i].replace(diff,1111111)
```

```python
# Reduce Validation levels, keep top 290 categories - based on train data
for i in np.r_[3,4,6,7,9,12,15]:
    count = pd.DataFrame(xval_df.iloc[:, i].value_counts())
    print(np.sum(count[:290]))
    wh_290 = xval_df.iloc[:,i].isin(count[:290].index)
    xval_df.loc[wh_290 == False, i] = 1111111
```

```python
for i in range(0, 21):
        print(i, len(xval_df.iloc[:, i].unique()) )
```

```python
# Combine levels for selected variables - Validation
for i in np.r_[3,4,6,7,9,12,15,18]:
    category = xtrain_df.iloc[:, i].unique()
    allcategory =xval_df.iloc[:,i].unique()
    diff = np.setdiff1d(allcategory, category)
    xval_df.iloc[: ,i] = xval_df.iloc[:,i].replace(diff, 1111111)
```

```python
#Change to category and same levels
for i in range(0,21):
    categories = np.union1d(xtrain_df.iloc[:,i], xval_df.iloc[:,i])
    xtrain_df.iloc[:,i] = pd.Categorical(xtrain_df.iloc[:,i], categories=categories)
    xval_df.iloc[:,i] = pd.Categorical(xval_df.iloc[:,i], categories=categories)
```

## Logistic - Ordinal Value

```python
# Fit Logistic Regression and Calculate Log Loss
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import log_loss
l = LogisticRegression()

ytrain=ytrain.astype('int')
yval=yval.astype('int')

l.fit(xtrain_label, ytrain)
ypred = l.predict(xval_label)
ypred_prob = l.predict_proba(xval_label)
print(log_loss(yval, ypred_prob))
```

```python
# Performance Measurement
from sklearn.metrics import accuracy_score, precision_score, recall_score, classification_report
from sklearn.metrics import confusion_matrix

accuracy = sum(ypred == yval)/len(yval)
error = 1 - accuracy
print("The predictive accuracy is: ", round(accuracy, 2))
print("The classification error is: ", round(error, 2))
print(classification_report(yval, ypred))

# Confusion matrix

cnf_matrix = pd.DataFrame(confusion_matrix(yval, ypred))
print("The Confusion matrix: \n", cnf_matrix)
```
